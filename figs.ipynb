```python
def plot_dyn(xx,yy,axs,ls,leg,lo):
    axs.plot(xx,yy,ls,label=leg)
    axs.set_box_aspect(0.618)
    axs.legend(loc=lo,frameon=False,ncol=2,fontsize=11)
    axs.set_xlim(0,300)
    axs.set_ylim(-1,1)
```

```python
def plot_fun(xx,yy,axs,ls,leg,lo):
    axs.plot(xx,yy,ls,label=leg)
    axs.set_box_aspect(0.618)
    axs.legend(loc=lo,frameon=False,fontsize=11)
    axs.set_xlim(0.0,0.5)
    #axs.set_ylim(0,6.8)
```

```python
#(in the first version of multi-D1 results,lambda_q for discrete modes is used)
Nmt=300
wc=1.0
kn=np.arange(0,Nmt+1)
xk=-wc*np.log(1-kn/Nmt)
wk=-wc*np.log(1-(kn[1:]-0.5)/Nmt)
gk10=np.zeros((Nmt))
gk30=np.zeros((Nmt))
gk50=np.zeros((Nmt))

def jwfun(x,s,alpha,wc):
    return 2*alpha*x**s*np.exp(-x**2/wc**2)


for jj in range(0,Nmt):
    gk10[jj]=sy.integrate.quad(jwfun,xk[jj],xk[jj+1],args=(1.0,0.1,wc))[0]
    gk30[jj]=sy.integrate.quad(jwfun,xk[jj],xk[jj+1],args=(1.0,0.3,wc))[0]
    gk50[jj]=sy.integrate.quad(jwfun,xk[jj],xk[jj+1],args=(1.0,0.5,wc))[0]
```
